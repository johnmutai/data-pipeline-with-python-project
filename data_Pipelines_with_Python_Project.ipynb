# Data Pipelines with Python Project

## Data Extraction, Cleaning and Transformation

```python
import pandas as pd

# Extract data

# Load data from dataset1 CSV file
dataset1 = pd.read_csv('dataset1.csv')

# Check for missing values
dataset1.isna().sum()

# Deal with missing values
dataset1.fillna('No Promo Code', inplace=True)

dataset1.rename(columns = {'country_of_purchase':'country'}, inplace = True)

```

```python
# Load data from dataset2 CSV file
dataset2 = pd.read_csv('dataset2.csv')

# Check for missing values
dataset2.isna().sum()

# Deal with missing values
#dataset1.fillna('No Promo Code', inplace=True)

dataset2.rename(columns = {'country_of_payment':'country'}, inplace = True)

```

```python
# Load data from dataset3 CSV file
dataset3 = pd.read_csv('dataset3.csv')

# Check for missing values
# dataset3.isna().sum()

dataset3.rename(columns = {'country_of_refund':'country'}, inplace = True)

```

## Merge the data

```python
# merge the data sets

merged_dataset = pd.merge(pd.merge(dataset1, dataset2, on=['customer_id', 'country', 'payment_method', 'payment_status']), dataset3, on=['customer_id', 'country'])

merged_dataset
```

## Load the data 

```python
merged_dataset.to_csv('customer_data.csv', index=False)
```
